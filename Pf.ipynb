```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("C:/Users/adile/Documents/Datos estructurados y estimados de locación/Proyecto_final/Embarazo legal.csv") 
# Mostrar las primeras filas del dataset
df.head()
```

#### Limpieza de datos:

```python
#Eliminar filas con valores nulos (NaN)
df = df.dropna()
#Eliminar filas duplicadas
df = df.drop_duplicates()
#Renombrar columnas si es necesario
df = df.rename(columns={'numero_hijos': 'n_hijos'})
df = df.rename(columns={'numero_embarazos': 'n_embarazos'})
df = df.rename(columns={'numero_abortos': 'n_abortos'})
df = df.rename(columns={'numero_iles': 'n_iles'})

#Cambiar tipos de datos de columnas 
df['edad_inicio_vida_sexual'] = df['edad_inicio_vida_sexual'].astype(int) 
df['semanas_embarazo'] = df['semanas_embarazo'].astype(int)
df['n_hijos'] = df['n_hijos'].astype(int)
df['n_embarazos'] = df['n_embarazos'].astype(int) 
df['n_abortos'] = df['n_abortos'].astype(int) 
df['n_iles'] = df['n_iles'].astype(int) 

#Resetear el índice después de la limpieza
df = df.reset_index(drop=True)

#Guardar el dataset limpio
df.to_csv('C:/Users/adile/Documents/Datos estructurados y estimados de locación/Proyecto_final/Embarazo legal.csv', index=False)

print("El dataset limpio ha sido guardado correctamente.")
df.head()
```

```python
# Calcular la media
promedio_edad = df['edad'].mean()
print(f'La edad promedio de las mujeres que abortan legalmete en México es de: {promedio_edad:,.2f} años')

# Calcular la mediana
mediana_edad = df['edad'].median()
print(f'El valor central de las mujeres que abortan legalmete en México es de: {mediana_edad:,.0f} años')

# Calcular la moda
moda_edad = df['edad'].mode()[0]
print(f'La edad más común de las mujeres que abortan legalmete en México es de: {moda_edad} años')
```

```python
#varianza

# Calcular la varianza de la edad
varianza_edad = df['edad'].var()
print(f'La varianza de las mujeres que abortan legalmete en México es de: {varianza_edad:,.2f}')
```

### Frecuencias

```python
#Rango
rango = df['edad_inicio_vida_sexual'].max() - df['edad_inicio_vida_sexual'].min()
#Número de clases.
n_clases = 1 + 3.322 * (df['edad_inicio_vida_sexual'].count() ** (1/3)) # Fórmula de Sturges.
#Tamaño de las clases.
tamaño_de_las_clases = rango / n_clases
#Definir los limites de los intervalos de clase.
limites = [df['edad_inicio_vida_sexual'].min() + i * tamaño_de_las_clases for i in range(int(n_clases) + 1)]
#Definir puntos medios de la clase
puntos_medios = [(limites[i] + limites[i + 1]) / 2 for i in range(int(n_clases))]

```

```python
#tabla de frecuencias 
tabla_frecuencias_nembarazos = pd.DataFrame({
    'Intervalo_de_clase': [f'({round(limites[i], 2)} - {round(limites[i + 1], 2)}]' for i in range(int(n_clases))],
    'Marca_de_clase': [round(puntos_medios[i], 2) for i in range(int(n_clases))],
    'Frecuencia_absoluta': [((df['edad_inicio_vida_sexual'] >= limites[i]) & (df['edad_inicio_vida_sexual'] < limites[i + 1])).sum() for i in range(int(n_clases))],
    'Frecuencia_relativa': [((df['edad_inicio_vida_sexual'] >= limites[i]) & (df['edad_inicio_vida_sexual'] < limites[i + 1])).sum() / df['edad_inicio_vida_sexual'].count() for i in range(int(n_clases))]
})

#Mostrar la tabla de frecuencias.
tabla_frecuencias_nembarazos.head().sort_values(by='Frecuencia_absoluta', ascending=False)
```

```python
#Graficar el histograma 
plt.figure(figsize=(10,6))
plt.hist(df['edad_inicio_vida_sexual'], bins= 10, color='purple', edgecolor='black', alpha=1) # 32 intervalos puedes modificarlos para ver diferentes distribuciones.

#Configurar los títulos y las etiquetas
plt.title('Histograma edad incio de vida sexual')
plt.xlabel('edad_inicio_vida_sexual')
plt.ylabel('Frecuencia')
plt.grid(axis='y')
plt.xticks(rotation=45)

#Mostrar el histograma
plt.show()
```
